2. assessments.csv

```python
# assessments.csv
assessments = pd.read_csv('./anonymisedData/assessments.csv')
assessments.head(10)
```

```python
assessments.info()
```

### assessments 课程评估方式表   
+ code_module 某课程的代码名称，作为标识符。一共有AAA-GGG 7种。  
+ code_presentation 某次课程开始的时间。它由年份和 "B "组成，"B "代表2月开始，"J "代表10月开始。  
+ id_assessment 评估的标识号  
+ assessment_type 存在三种类型的评估：老师分数评估（TMA），计算机分数评估（CMA）和期末考试（Exam）。  
+ date 有关评估的最终提交日期的信息，该信息以课程开始以来的天数计算。课程的开始日期为数字0（零）。  
+ weight 评估的权重，以％为单位。通常，考试要单独处理，权重为100％；所有其他评估的总和为100％。

4.studentInfo.csv

```python
studentInfo = pd.read_csv('./anonymisedData/studentInfo.csv')
studentInfo.head(10)
```

```python
studentInfo.info()
```

### studentInfo 学生信息表    
+ code_module 某课程的代码名称，作为标识符。一共有AAA-GGG 7种。  
+ code_presentation 某次课程开始的时间。它由年份和 "B "组成，"B "代表2月开始，"J "代表10月开始。

+ most_education 加入课程时的最高教育水平

+ age_band 学生的年龄范围
+ num_of_prev_attempts 学生曾经学习该课程的次数
+ students_credits 学生当前正在学习的课程的学分总数  
+ disability 指示学生是否残疾  
+ final_result 学生在课程中的最终结果

5.studentRegistration.csv

```python
studentRegistration = pd.read_csv('./anonymisedData/studentRegistration.csv')
studentRegistration.head(10)
```

```python
studentRegistration.info()
```

### studentRegistration 学生注册表  
+ code_module 某课程的代码名称，作为标识符。一共有AAA-GGG 7种。  
+ code_presentation 某次课程开始的时间。它由年份和 "B "组成，"B "代表2月开始，"J "代表10月开始。  
+ id_student 学生的唯一标识号  
+ date_registration 学生在课程上注册的日期，这是相对于课程开始的天数（-30表示学生在课程开始前30天注册了该日期）。  
+ date_unregistration 从课程中注销学生的日期，这是相对于课程开始的天数。完成课程的学生将此字段留空。未注册的学生将 Withdrawal作为studentInfo.csv文件中final_result列的值。

### studentAssessment 学生评价表  
+ id_assessment 评估的标识号
